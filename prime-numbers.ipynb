```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt

# Array of numbers 0-100
numbers = np.arange(100)

# List for prime numbers
primes = []

# Prime number flag
is_prime = True 

# Find prime numbers
for i in numbers:
    if((i != 0) & (i != 1)):
        for j in np.arange(i):
            if((j != 0) & (j != 1)):
                if(i % j == 0):
                    #print(str(i) + ": is not a prime")
                    is_prime = False
                    break
        if(is_prime == True):
            primes.append(float(i))
    is_prime = True
    
    
print("Primes:")    
print(primes)        

# Insert in array and plot
y = np.asarray(primes)
x = np.arange(len(primes))

plt.axes(projection = 'polar')
plt.plot(x,y)
plt.show()
```
